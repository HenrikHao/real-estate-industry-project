# Processing Census Data

```python
import pandas as pd

df_2021 = pd.read_csv("../data/landing/census2021/2021 Census GCP All Geographies for VIC/SA2/VIC/2021Census_G02_VIC_SA2.csv")
df_2021
```

```python
df_2016 = pd.read_csv("../data/landing/census2016/2016 Census GCP All Geographies for VIC/SA2/VIC/2016Census_G02_VIC_SA2.csv")
df_2016
```

```python
corresponding_table = pd.read_csv("../data/landing/CG_SA2_2016_SA2_2021.csv")
corresponding_table.head()
```

```python
# Merge the 2016 census data with the corresponding table
merged_2016 = pd.merge(df_2016, corresponding_table[['SA2_MAINCODE_2016', 'SA2_CODE_2021']], on='SA2_MAINCODE_2016', how='left')

# Drop the old SA2_MAINCODE_2016 column
merged_2016.drop('SA2_MAINCODE_2016', axis=1, inplace=True)

merged_2016
```

```python
df_2016_cleaned = merged_2016.drop_duplicates(subset='SA2_CODE_2021', keep='first')

# Check the shape of the cleaned dataframe
print(df_2016_cleaned.shape)
```

```python
import pandas as pd


# Filter columns
df_2021 = df_2021[['SA2_CODE_2021', 'Median_tot_prsnl_inc_weekly', 'Median_rent_weekly']]
df_2016 = df_2016_cleaned[['SA2_CODE_2021', 'Median_tot_prsnl_inc_weekly', 'Median_rent_weekly']]

df_2021['SA2_CODE_2021'] = df_2021['SA2_CODE_2021'].astype(int)
df_2016['SA2_CODE_2021'] = df_2016['SA2_CODE_2021'].astype(int)

# Merge dataframes
merged = df_2021.merge(df_2016, on='SA2_CODE_2021', suffixes=('_2021', '_2016'))

# Calculate annual growth rate
merged['growth_rate_prsnl_inc'] = (merged['Median_tot_prsnl_inc_weekly_2021'] / merged['Median_tot_prsnl_inc_weekly_2016']) ** (1/5) - 1
merged['growth_rate_rent'] = (merged['Median_rent_weekly_2021'] / merged['Median_rent_weekly_2016']) ** (1/5) - 1

# Create a new dataframe to store the long-format data
long_df = pd.DataFrame(columns=['year', 'SA2_CODE_2021', 'Median_tot_prsnl_inc_weekly', 'Median_rent_weekly'])

# Fill out the data for the years 2016 to 2021
for year in range(2016, 2022):
    if year == 2016:
        temp_df = merged[['SA2_CODE_2021', 'Median_tot_prsnl_inc_weekly_2016', 'Median_rent_weekly_2016']]
        temp_df.columns = ['SA2_CODE_2021', 'Median_tot_prsnl_inc_weekly', 'Median_rent_weekly']
    elif year == 2021:
        temp_df = merged[['SA2_CODE_2021', 'Median_tot_prsnl_inc_weekly_2021', 'Median_rent_weekly_2021']]
        temp_df.columns = ['SA2_CODE_2021', 'Median_tot_prsnl_inc_weekly', 'Median_rent_weekly']
    else:
        prev_year = year - 1
        temp_df = long_df[long_df['year'] == prev_year].copy()
        
        # Use .map() to match the SA2_CODE_2021 values and apply the growth rate
        temp_df['Median_tot_prsnl_inc_weekly'] = temp_df['Median_tot_prsnl_inc_weekly'] * (1 + temp_df['SA2_CODE_2021'].map(merged.set_index('SA2_CODE_2021')['growth_rate_prsnl_inc']))
        temp_df['Median_rent_weekly'] = temp_df['Median_rent_weekly'] * (1 + temp_df['SA2_CODE_2021'].map(merged.set_index('SA2_CODE_2021')['growth_rate_rent']))
    
    temp_df['year'] = year
    long_df = pd.concat([long_df, temp_df], ignore_index=True)

# Print the long-format dataframe
print(long_df)
```

```python
import pandas as pd
import numpy as np

# 1. Rename columns in data_median for consistency
data_median = data_median.rename(columns={'cost': 'Median_rent_weekly', 'SA2_CODE21': 'SA2_CODE_2021'})

# 2. Filter long_df to keep only the SA2 codes present in data_median
filtered_long_df = long_df[long_df['SA2_CODE_2021'].isin(data_median['SA2_CODE_2021'])]

# 3. Calculate the growth rate for Median_tot_prsnl_inc_weekly from 2016 to 2021
growth_rates = filtered_long_df.groupby('SA2_CODE_2021').apply(lambda group: group['Median_tot_prsnl_inc_weekly'].iloc[-1] / group['Median_tot_prsnl_inc_weekly'].iloc[0]) ** (1/5) - 1

# 4. Use the growth rate to estimate Median_tot_prsnl_inc_weekly for 2022 and 2023
for year in [2022, 2023]:
    last_year_data = filtered_long_df[filtered_long_df['year'] == year - 1].set_index('SA2_CODE_2021')
    new_data = last_year_data.copy()
    new_data['year'] = year
    new_data['Median_tot_prsnl_inc_weekly'] = np.nan
    filtered_long_df = pd.concat([filtered_long_df, new_data.reset_index()])

# 5. Estimate Median_rent_weekly for 2022 using the average of 2021 and 2023 data
data_2021 = filtered_long_df[filtered_long_df['year'] == 2021].set_index('SA2_CODE_2021')['Median_rent_weekly']
data_2023 = data_median.set_index('SA2_CODE_2021')['Median_rent_weekly']
data_2022 = (data_2021 + data_2023) / 2
filtered_long_df = filtered_long_df.reset_index(drop=True)
filtered_long_df.loc[filtered_long_df['year'] == 2022, 'Median_rent_weekly'] = filtered_long_df['SA2_CODE_2021'].map(data_2022)

# 6. Extract constant columns from data_median and merge with filtered_long_df
constant_cols = ['beds', 'baths', 'parkings', 'Nearest_station', 'Nearest_park', 'Nearest_shop', 'Nearest_hospital', 'Nearest_school', 'Nearest_supermarket']
constant_data = data_median[['SA2_CODE_2021'] + constant_cols]
result = pd.merge(filtered_long_df, constant_data, on='SA2_CODE_2021', how='left')

result.head()
```
