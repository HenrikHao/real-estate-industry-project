```python
import pandas as pd
```

```python
data = pd.read_csv("../data/aggregated_dataset.csv")
```

```python
data_median = data.groupby('SA2_NAME21').median().reset_index()
```

```python
data_median
```

```python
crime = pd.read_csv("../data/raw/Offence count by postcode.csv")
```

```python
crime_median = crime.groupby('SA2_NAME21').median().reset_index()
```

```python
aggregated = data_median.merge(crime_median[['SA2_NAME21', 'Offence Count']], on='SA2_NAME21', how='left')
```

```python
aggregated
```

```python
population_forecast = pd.read_csv("../data/external_SA2/Victoria.csv")
```

```python
population_forecast
```

```python
population_forecast = population_forecast.groupby(['YEAR', 'SA2_NAME']).sum().reset_index()
```

```python
population_forecast = population_forecast.rename(columns={'SA2_NAME': 'SA2_NAME21'})
```

```python
population_forecast_pivot = population_forecast.pivot(index='SA2_NAME21', columns='YEAR', values='Total')
population_forecast_pivot.reset_index(inplace=True)
```

```python
population = population_forecast_pivot.iloc[:, 0:11]
```

```python
population
```

```python
merged = aggregated.merge(population, on='SA2_NAME21', how='left')
```

```python
column_mapping = {year: f'population_{year}' for year in range(2017, 2027)}
merged.rename(columns=column_mapping, inplace=True)
```

```python
merged
```

```python
census = pd.read_csv("../data/landing/census2021/2021 Census GCP All Geographies for VIC/SA2/VIC/2021Census_G02_VIC_SA2.csv")
```

```python
census = census.rename(columns={'SA2_CODE_2021': 'SA2_CODE21'})
```

```python
all_features = merged.merge(census, on='SA2_CODE21', how='left')
```

```python
all_features.to_csv("../data/raw/Properties with distance, crime, population, affluence.csv")
```
