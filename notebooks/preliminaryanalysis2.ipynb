```python
# Randomly sample the data
sample_size = int(len(filtered_df) * 0.1)
sampled_df = filtered_df.sample(n = sample_size, random_state=42)
sampled_df
```

```python
# Visualize the distribution of each variable(beds; baths; parkings; weekly_rent) using histograms.

def draw_histogram(df, column_name, subplot):
    plt.subplot(subplot)
    sns.histplot(data = df, x = column_name, kde = True, bins = 10, edgecolor='black')
    plt.xlabel(column_name)
    plt.ylabel('Frequency')
    plt.title(f'{column_name} Distribution Histogram')

plt.figure(figsize=(12, 8))
draw_histogram(sampled_df, 'beds', 221)
draw_histogram(sampled_df, 'baths', 222)
draw_histogram(sampled_df, 'parkings', 223)
draw_histogram(sampled_df, 'weekly_rent', 224)
plt.tight_layout()  
plt.show()
```

```python
def relationship_histogram(df, column_name, subplot):
    plt.subplot(subplot)
    sns.histplot(data=df, x='weekly_rent', hue=column_name, bins=10, kde=True)
    plt.xlabel('Weekly Rent')
    plt.ylabel(column_name)
    plt.title(f'Weekly Rent Histogram by {column_name}')

plt.figure(figsize=(16, 8))
relationship_histogram(sampled_df, 'beds', 131)
relationship_histogram(sampled_df, 'baths', 132)
relationship_histogram(sampled_df, 'parkings', 133)
plt.tight_layout()  
plt.show()    
```

```python
# Create a pair plot for the selected columns
sns.pairplot(data=sampled_df[columns_to_check], kind='reg', diag_kind='kde')
plt.show()
```

```python
# Visualizing the distribution of 'weekly_rent' for different categories of 'beds', 'baths', and 'parkings' simultaneously --- Violin Plots
plt.figure(figsize=(10, 5))

plt.subplot(131)
sns.violinplot(x='beds', y='weekly_rent', data=sampled_df)
plt.title('Bedrooms vs. Weekly Rent')

plt.subplot(132)
sns.violinplot(x='baths', y='weekly_rent', data=sampled_df)
plt.title('Bathrooms vs. Weekly Rent')

plt.subplot(133)
sns.violinplot(x='parkings', y='weekly_rent', data=sampled_df)
plt.title('Parkings vs. Weekly Rent')

plt.tight_layout()
plt.show()
# White dots represent the median; the wider the plot, the higher density of weekly rent values in that range
```

```python
# To visualize the distribution of 'weekly_rent' for different combinations of 'beds', 'baths', and 'parkings' --- Pairwise Distribution Plots

g = sns.FacetGrid(sampled_df, col='beds', row='baths', margin_titles=True)
g.map_dataframe(sns.histplot, x='weekly_rent', bins=10)
g.set_axis_labels('Weekly Rent', 'Count')
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Weekly Rent Distribution by Bedrooms and Bathrooms')
plt.show()
```
