```python
import pandas as pd
properties_df = pd.read_csv("../data/raw/properties.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Plot the correlation metric between the weather feature and the demand
CORR_COLS = ['beds', 'baths', 'parkings', 'weekly_rent']

correlation_matrix = properties_df[CORR_COLS].corr()

sns.heatmap(correlation_matrix, annot=True, fmt=".2f") 

plt.title('Pearson Correlation Metric')

plt.savefig('../plots/property_corr.png', dpi=300, bbox_inches='tight')
```

```python
sns.boxplot(data=properties_df, x="beds", y="weekly_rent")
plt.title('Weekly Rental Price by Number of Beds')
```
