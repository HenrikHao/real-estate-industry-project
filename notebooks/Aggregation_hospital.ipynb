```python
import geopandas as gpd
import pandas as pd
```

```python
victoria_gdf = gpd.read_file('../data/Statistical_area_level2/victoria.geojson')
```

```python
hospital_gdf = gpd.read_file('../data/external_SA2/Hospital.geojson')
```

```python
joined_hospital = gpd.sjoin(hospital_gdf, victoria_gdf, how="inner", op="within")
```

```python
joined_hospital.columns
```

```python
hospital_SA2 = joined_hospital[['geometry', 'SA2_CODE21', 'SA2_NAME21']]
```

```python
hospital_SA2 = hospital_SA2.rename(columns={'geometry': 'Hospital_geometry', 'SA2_CODE21': 'Hospital_SA2_Code', 'SA2_NAME21': 'Hospital_SA2_Name'})
```

```python
school_df = pd.read_csv("../data/external_SA2/complete_school_location.csv", encoding='latin-1')
school_df['coordinates'] = school_df.apply(lambda row: [row['Y'], row['X']], axis=1)
```

```python
import geopandas as gpd
from shapely.geometry import Point

# Your property data
school_metadata = school_df

# Convert the coordinates column to a suitable format (assuming it's a string like '[lat, lon]')
#school_metadata['coordinates'] = school_metadata['coordinates'].apply(eval)

# Create points and names from the property_metadata
coords = [Point(xy[1], xy[0]) for xy in school_metadata['coordinates']]
names = school_metadata['School_Type'].tolist()

# Convert to GeoDataFrame
school_gdf = gpd.GeoDataFrame({'name': names, 'geometry': coords})
```

```python
joined_school = gpd.sjoin(school_gdf, victoria_gdf, how="inner", op="within")
```

```python
school_SA2 = joined_school[['name', 'geometry', 'SA2_CODE21', 'SA2_NAME21']]
school_SA2 = school_SA2.rename(columns={'name':'School_Type', 'geometry': 'School_Geometry', 'SA2_CODE21	': 'School_SA2_Code', 'SA2_NAME21': 'School_SA2_Name'})
```
