```python
from openpyxl import load_workbook
import pandas as pd
```

```python
wb=load_workbook('../data/external_SA2/homeless_number.xlsx')
homeless_sheet = wb['Table_5.3']
```

```python
data = homeless_sheet.values
columns = next(data)
df = pd.DataFrame(data, columns=columns)
```

```python
df
```

```python
new_column_names = df.iloc[4]
df.set_axis(new_column_names, axis=1, inplace=True)
df = df[4:]
```

```python
df
```

```python
homeless_df = df.iloc[1: ,[0, 3, 4]]
```

```python
homeless_df.rename(columns={"no.": "homeless number", df.columns[0]: 'ABS'}, inplace=True)
homeless_df.reset_index(drop=True, inplace=True)
```

```python
homeless_df
```

```python
homeless_df['ABS'].unique()
# we want the data between Victoria and Queensland
```

```python
row_index1 = homeless_df.loc[homeless_df["ABS"] == "Victoria"].index[0]
row_index2 = homeless_df.loc[homeless_df["ABS"] == "Queensland"].index[0]
```

```python
Vic_homeless = homeless_df.iloc[row_index1: row_index2]
```

```python
Vic_homeless
```

```python
Vic_homeless['SA2'].isna().sum() # only 84, can delete directly
```

```python
Vic_homeless['homeless number'].isna().sum()
```

```python
Vic_homeless.dropna(subset=['SA2'], inplace=True)
```
