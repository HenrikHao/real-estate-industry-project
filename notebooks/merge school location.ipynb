```python
import pandas as pd
import geopandas as gpd
```

```python
school_external = pd.read_csv('../data/landing/school_location.csv', encoding='latin-1')
```

```python
school_external.head()
```

```python
school_external.columns
```

```python
school_external = school_external[['School_Name','School_Type', 'Postal_Town', 'X', 'Y']]
```

```python
school_external.drop_duplicates()
```

```python
school_external['School_Type'].unique()
```

```python
school_external = school_external[school_external['School_Type'].isin(['Primary', 'Pri/Sec', 'Secondary', 'Language'])]
```

```python
len(school_external)
```

### Now there are 2182 schools, including primary, secondary and language schools
### Next, we consider adding university locations

```python
file_path = '../data/landing/Uni & College.geojson'
university_data = gpd.read_file(file_path)
```

```python
university_data.columns
```

```python
university_data = university_data[['amenity', 'name', 'geometry']]
```

```python
university_data['X'] = university_data['geometry'].x
university_data['Y'] = university_data['geometry'].y
university_data['School_Type'] = university_data['amenity']
university_data['School_Name'] = university_data['name']
university_data = university_data.drop(columns='amenity').drop(columns='name')
```

### There are 75 instances about university and college

```python
university_data
```

```python
school_combined = pd.concat([university_data, school_external], axis=0)
```

### In total, there are 2257 instances about school (Secondary, Primary, Language, University, College) locations

```python
school_combined.drop_duplicates()
```

```python
school_combined.to_csv('../data/raw/school_location.csv', index=False) 
```
