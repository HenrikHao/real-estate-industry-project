```python
from openpyxl import load_workbook
import pandas as pd
```

```python
wb=load_workbook('../data/external_SA2/homeless_number.xlsx')
homeless_2021 = wb['Table_5.3']
homeless_2016 = wb['Table_5.4']
```

```python
def read_sheet(sheet):
    data = sheet.values
    columns = next(data)
    return pd.DataFrame(data, columns=columns)

df_2021 = read_sheet(homeless_2021)
df_2016 = read_sheet(homeless_2016)
```

```python
# This function is to rename headers and slice df
def format_df(df):
    new_column_names = df.iloc[4]
    df.set_axis(new_column_names, axis=1, inplace=True)
    df = df[4:]
    df = df.iloc[1: ,[0, 3, 4]]
    df.rename(columns={"no.": "homeless number", df.columns[0]: 'ABS'}, inplace=True)
    df.reset_index(drop=True, inplace=True)
    return df

df_2021 = format_df(df_2021)
df_2016 = format_df(df_2016)
```

```python
# This function is to find Vic homeless number
def find_vic_homeless(df):
    row_index1 = df.loc[df["ABS"] == "Victoria"].index[0]
    row_index2 = df.loc[df["ABS"] == "Queensland"].index[0]
    vic_homeless = df.iloc[row_index1: row_index2]
    return vic_homeless

vic_homeless_2016 = find_vic_homeless(df_2016)
vic_homeless_2021 = find_vic_homeless(df_2021)


```
