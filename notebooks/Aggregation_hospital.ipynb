```python
import geopandas as gpd
import pandas as pd
```

```python
victoria_gdf = gpd.read_file('../data/Statistical_area_level2/victoria.geojson')
```

```python
hospital_gdf = gpd.read_file('../data/external_SA2/Hospital.geojson')
```

```python
hospital_gdf
```

```python
joined_gdf = gpd.sjoin(hospital_gdf, victoria_gdf, how="inner", op="within")
```

```python
joined_gdf.columns
```

```python
hospital_SA2 = joined_gdf[['geometry', 'SA2_CODE21', 'SA2_NAME21']]
```

```python
hospital_SA2.rename(columns={'geometry': 'Hospital_geometry', 'SA2_CODE21': 'Hospital_SA2_Code', 'SA2_NAME21': 'Hospital_SA2_Name'})
```

```python
hospital_SA2.to_csv('../data/raw/hospital_SA2.csv', index=False) 
```
